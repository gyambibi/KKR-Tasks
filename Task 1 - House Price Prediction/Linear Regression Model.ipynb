```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
df_train = pd.read_csv('train.csv')
df_train
```

```python
numeric_df_train = df_train.select_dtypes(include=[np.number])
numeric_df_train
```

```python
x_train = numeric_df_train.iloc[:, :-1].values
```

```python
y_train = numeric_df_train.iloc[:, -1].values
```

```python
plt.scatter(x_train[:, 0], y_train)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
df_test = pd.read_csv('test.csv')
numeric_df_test = df_test.select_dtypes(include=[np.number])
x_test = numeric_df_test.iloc[:, :].values
numeric_df_test
```

```python
y_pred = model.predict(x_test)
y_pred
```
